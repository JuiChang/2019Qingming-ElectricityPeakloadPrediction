#### Goal  
預測 2019/4/2 ~ 2019/4/8 的每日"電力尖峰負載"(MW)
* NCKU: spring break - 3/30 ~ 4/7  
* Qingming vacation: 4/4 ~ 4/7

#### Metric  
作業將以 “尖峰負載預測值” 與 "尖峰負載實際數值"之 Root-Mean-Squared-Error (RMSE) 作為評估分數。

#### Data

* [氣象局觀測資料查詢](https://e-service.cwb.gov.tw/HistoryDataQuery/index.jsp)：2017~2019/3/31之台中觀測站下午一點之氣溫
* Google搜尋：“一週天氣 台中”：2019/4/1~2019/4/2之台中氣溫預測 

#### Load data and create dummy variable for weekday

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1718 = pd.read_csv('2017_19326_t.csv')
# change to common datetime format, and set as index
data1718['日期'] = pd.to_datetime(data1718['日期'], format='%Y%m%d')
data1718.set_index('日期', inplace=True)

# add weekday column and move to front
data1718['weekday'] = data1718.index.dayofweek
cols = data1718.columns.tolist()
cols = cols[-1:] + cols[:-1]
data1718 = data1718[cols]
```

### Exploration
* plot daily peakload
* mean peakload of each month 
* mean peakload of Monday to Sunday
* peakload around Qingming

#### daily peakload

```python
# extract '尖峰負載(MW)' as a Series
peakload = data1718.loc[:, '尖峰負載(MW)']

from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
plt.xticks(rotation=70)
plt.plot(peakload)
```

* Weeday peakload and weekend peakload rise and fall together through out the seasons (or say, months).  
* The two extremely low positions are Chinese New Year Vacation
    * 2017/1/27 ~ 2/1 Chinese New Year  
    * 2018/2/15 ~ 2/20 Chinese New Year 
* Roughly, there are three low positions in each year, they are:
    * New Year
    * Chinese New Year  
    * Qingming  
* Roughly, peakload are rising after Qingming.  

#### mean peakload of each month

```python
yearmonth = peakload.index.map(lambda x: 100*x.year + x.month)
pl_eachmonth_mean = peakload.groupby(yearmonth).mean()
pl_eachmonth_mean.index = (pd.to_datetime(pl_eachmonth_mean.index, format='%Y%m'))

plt.subplots_adjust(hspace = 1)
plt.subplot(211)
color = 'tab:blue'
plt.title('mean peakload of each month')
plt.xticks(rotation=30)
plt.plot(pl_eachmonth_mean, '-ok', color=color)

pl_month_mean = peakload.groupby(peakload.index.month).mean()

plt.subplot(212)
color = 'tab:blue'
plt.title('(combine same months among different years)')
plt.plot(pl_month_mean, '-ok', color=color)
```

* 5~11(month): high peakload
* Roughly, 2018 > 2017 in the first half of the years, while 2017 > 2018 of the second half of the years

#### mean peakload of Monday to Sunday

```python
pl_weekday_mean = peakload.groupby(data1718['weekday']).mean()

color = 'tab:blue'
plt.title('mean peakload of each weekday')
plt.xlabel('weekday (Mon. to Sun.)')
plt.plot(pl_weekday_mean, '-ok', color=color)
```

* Obviously, weekends have lower peakload, and Sunday is lower than Saturday.  
* Monday is the lowest in weekdays.

### Our predicting target include 2019 spring break, so let's explore the same periods of 2017 and 2018  
#### 2017 around spring break    
NCKU: spring break - 3/30 ~ 4/4  
Qingming vacation: 4/1 ~ 4/4  

#### 2018 around spring break    
NCKU: spring break - 3/31 ~ 4/8  
Qingming vacation: 4/4 ~ 4/8  

```python
import datetime as dt
# 2017 around spring break, Qingming(4/1~4/4)
start_date = dt.datetime(2017,3,1)
end_date = dt.datetime(2017,4,30)
mask17 = (peakload.index > start_date) & (peakload.index <= end_date)

fig, ax1 = plt.subplots()
color = 'tab:blue'
plt.xticks(rotation=90)
plt.title('2017 around spring break, Qingming(4/1~4/4, Sat~Tue) peakload')
plt.plot(peakload.loc[mask17], '-ok', color=color)

ax2 = ax1.twinx()
color = 'tab:red'
plt.plot(data1718.loc[mask17]['weekday'], 'ro', color=color)

# 2018 around spring break, Qingming(4/4~4/8)
start_date = dt.datetime(2018,3,1)
end_date = dt.datetime(2018,4,30)
mask18 = (peakload.index > start_date) & (peakload.index <= end_date)

fig, ax1 = plt.subplots()
color = 'tab:blue'
plt.xticks(rotation=90)
plt.title('2018 around spring break, Qingming(4/4~4/8, Wed~Sun) peakload')
plt.plot(peakload.loc[mask18], '-ok', color=color)

ax2 = ax1.twinx()
color = 'tab:red'
plt.plot(data1718.loc[mask18]['weekday'], 'ro', color=color)

# 2017 & 2018
fig, ax1 = plt.subplots()
color = 'tab:blue'
plt.xticks(rotation=90)
plt.title('2017 & 2018 around spring break peakload')
plt.tick_params(axis='x', labelcolor=color)
plt.plot(peakload.loc[mask17], '-ok', color=color)

ax2 = ax1.twiny()
color = 'tab:red'
plt.xticks(rotation=90)
# plt.title('2018 around spring break peakload')
plt.tick_params(axis='x', labelcolor=color)
plt.plot(peakload.loc[mask18], '-ok', color=color)
```

* University's spring break doesn't affect the peakload much, instead, the qingming vacation affect peakload apparently.  
* Peakload of Qingming vacation seems like the peakloads of weekends, but may even be lower.
* In this period, peakloads of 2018 are higher than peakloads of 2017 roughly. **We should consider the trend of 2019 when making our prediction**
* No matter are normal weekday or qingming vacation, peakload of today is related to the peakload yesterday, roughly. **We should consider peakload of yesterday when making our prediction**
* In Qingming vacations of 2017 & 2018, once the peakload rise, it won't fall again (i.e. **no 'W' shape**). So do the two Chinese New Years.
* Peakloads are roughly rising after Qingming vacation.
* The **first day after Qingming** act like normal Monday. The peakload is **slightly lower than recent normal weekday**.

### Models

#### For 4/2 ~ 4/8, fit one model for each day
* 4/2: xgboost model predicting peakload of Tuesday while the peakloads of previous several days are known.
* 4/3: xgboost model predicting peakload of Wednesday while the peakload of the previous Tuesday is unknown.
* 4/8: xgboost model predicting peakload of Monday while the peakloads of the previous Tuesday to Sunday are unknown.
* 4/4 ~ 4/7: Four linear regression models predicting the peakloads of the first to forth days of Qingming vacation seperately. 
    * Take the first vacation day for example : For predicting 2019/4/4, we fit a linear regression by two observations: 2017/4/1 and 2018/4/4, for each observation:
        * X: mean peakload of its previous five weekdays(exclude weekend), 氣溫 of that day.
        * Y: peakload of that day.

#### Feature selection for models
* Consider only 尖峰負載、備轉容量、氣溫, because for the period later than 2019/1/31, I can only get these data.
* For 4/2, 4/3 and 4/8, do seperately:
    * Cross validate how many pervious known day should we consider.
    * Cross validate which months of data should we consider. (motivation: For peakloads data after August of a year, the trend is much different with the trend around Qingming, so it's less useful as training data.)
    * Cross validate whether applying 備轉容量 as predictor.
    * (For the chosen values, please refer to code: predict.py)
* Tune the hyperparameters for the 4/2 xgb model, and apply them on the models of 4/3 and 4/8. ([Refer to the xgb tuning tutorial](https://www.analyticsvidhya.com/blog/2016/03/complete-guide-parameter-tuning-xgboost-with-codes-python/))
* Peakloads of weekdays and weekends are much different, thus, the training data of 4/2, 4/3 and 4/8 only include weekdays (Mon. to Fri.)
